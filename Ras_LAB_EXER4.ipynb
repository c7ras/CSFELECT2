```python
import pandas as pd
import numpy as np
df = pd.read_csv('convolution_model_training_history.csv', encoding='latin-1')
loss = df['loss']
vloss = df['val_loss']
plt.plot(range(0, epochs), loss, "skyblue", label='Train loss')
plt.plot(range(0, epochs), vloss, "orange", label='Valid loss')
plt.legend()
plt.show()
```

```python
traincce = df['mrcnn_class_loss']
validclassce = df['val_mrcnn_class_loss']

plt.plot(range(0,epochs),traincce,"skyblue", label='Train class ce')
plt.plot(range(0,epochs),validclassce,"orange", label='Valid class ce')

plt.legend()

plt.show
```

```python
tbloss = df['val_mrcnn_bbox_loss']
vbloss = df['mrcnn_bbox_loss']
plt.plot(range(0,epochs),tbloss,"orange",label='Train box loss')
plt.plot(range(0,epochs),vbloss,"skyblue",label='Valid box loss')
plt.legend()
plt.show
```

```python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.plot(range(0, epochs), loss, "skyblue", label='Train loss')
plt.plot(range(0, epochs), vloss, "orange", label='Valid loss')
plt.legend()

plt.subplot(132)
plt.plot(range(0,epochs),traincce,"skyblue", label='Train class ce')
plt.plot(range(0,epochs),validclassce,"orange", label='Valid class ce')
plt.legend()

plt.subplot(133)
plt.plot(range(0,epochs),tbloss,"orange",label='Train box loss')
plt.plot(range(0,epochs),vbloss,"skyblue",label='Valid box loss')
plt.legend()

plt.show()
```
